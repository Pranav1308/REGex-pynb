```python
fig = plt.figure(figsize = (6,6)) #intialization of figure
```

```python
p1 = fig.add_subplot(2,2,1)
p2 = fig.add_subplot(2,2,2)    #this command intializes subplots on the figure
p3 = fig.add_subplot(2,2,3)    #there should be n variables like p1,p2,p3,p4 because we have given 2,2 argument i.e along the row and along the column
p4 = fig.add_subplot(2,2,4)    #in short multiply first 2 arguments and there should be that much variables
#first argument is along row and second is along the column and the third one is kind of index to remembre the pllot no.
```

```python
p1.plot(np.arange(10), 'k-')  #here k is the black colour and - indicates simple line 
```

```python
p2.scatter(np.arange(30), np.arange(30) + np.random.randn(30))
```

```python
p3.plot(np.arange(10), 'm--') 
```

```python
p4.plot(np.arange(10), 'co-') 
```

```python
plt.figure()
x = np.arange(10)
y = np.arange(10)
plt.plot(x,y, 'r-*')
```

```python
plt.figure(figsize = (5,5))
plt.plot(np.random.randn(30), 'mo-', label = 'FirstPlot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple line plot')
plt.legend()
```

```python
plt.figure()
x = np.arange(10)
y = x ** 2
z = x ** 3
plt.plot(x,y, 'm*-', label = 'Squares')
plt.plot(x,z, 'bo-', label = 'Cubes')
plt.title('Squares and Cubes')
plt.legend()
```

```python
plt.figure()
x = np.arange(10)
y = x ** 2
plt.scatter(x,y, s = 100, c = 'm')# c is colour and s is size
```

```python
plt.xticks(np.arange(10))
```

```python
plt.yticks(np.arange(0,100,5))
```

```python
plt.figure()
x = np.arange(10)
y = x * 2
plt.bar(x,y, width = 0.8)
plt.xticks(x)
plt.yticks(y)
```

```python
plt.figure()
x = np.arange(10)
y = x * 2
z = x * 4
plt.bar(x,y, width = 0.4, color = 'm')
plt.bar(x,z, width = 0.4, color = 'orange', bottom = x)#if we dont write bottom  = x then the orange command will be shown only it will get stacked on each other to avodi that the bottom command is used
```

```python
plt.figure()
x = np.arange(10)
y = x * 2
z = x * 4
plt.bar(x-0.2,y, width = 0.4, color = 'm')
plt.bar(x+0.3,z, width = 0.4, color = 'c')
```
